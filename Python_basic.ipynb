2. what is an interpreter in pyhon?
   - In Python, an **interpreter** is a program that reads and executes your Python code line by line. Unlike compiled languages (like C or Java), which are converted entirely into machine code before running, Python uses an interpreter to execute code directly.

3. What are pre-defined keywords in python?  
  - **Pre-defined keywords** in Python are reserved words that have special meaning in the language. You cannot use them as variable names, function names, or identifiers, because Python uses them to define its syntax and structure.

4. Can keywords be used as variable names?  
   - **No**, keywords cannot be used as variable names in Python.

5. What is mutability in Python?
   - Mutable is when something is changeable or has the ability to change. In Python, **mutable** is the ability of objects to change their values. These are often the objects that store a collection of data.

6. Why are lists mutable, but tuples are immutable?
   - **Why Are Lists Mutable?**
      -  In Python, **lists** are designed to be flexible containers. They are meant to store and manipulate collections of data, so mutability is a feature, not a flaw.
   - **Tuple are immutable?**
       - They are often used as fixed data structures, where changes would break expectations
       - Immutability allows them to be used as dictionary keys or set elements.

7. What is the difference between "==" and "is" operators in Python?
   - **== (Equality Operator)**
      - Compares the values of two objects.
      - Returns True if the contents are the same.
   - **is (Identity Operator)**
       - Compares the identity (memory location) of two objects.
       - Returns True if both variables point to the same object in memory.

8. What are logical operators in Python?     
   - Logical operators in Python are used to combine conditional statements and return Boolean values (True or False). They're essential in control flow like if, while, and loops.
   There are 3 Logical Operators (and, or, not).

9. What is type casting in python?   
   - Type casting also known as type conversion in Python means changing the data type of a value or variable from one type to another.

10. What is the difference between implicit and explicit type casting?
    - **Implicit Type Casting**
       - Automatically done by Python itself without programmer intervention.
       - when Python converts one data type to another during operations to avoid errors.
       - Usually occurs when mixing types in expressions.
   - **Explicit Type Casting**
      - Done manually by the programmer using conversion functions like int(), float(), str(), etc.
      - You explicitly tell Python to convert a value from one type to another.

11. What is the purpose of conditional statements in Python?
    - **Conditional statements** let your program make decisions by executing different blocks of code based on conditions — basically, they help your program choose what to do next.

12. How does the elif statement work?
    - Its used to check multiple conditions sequentially after an initial if condition.
    - If the if condition is False, Python checks the elif condition.
    - You can have multiple elif statements to test different conditions.
    - Once a True condition is found, Python skips the rest of the elif and else blocks.
    - elif lets you test multiple, exclusive conditions clearly and efficiently.

13. What is the difference between for and while loops?
    - **for Loop**
       - Used when you want to iterate over a sequence (like a list, string, range, etc.).
       - Best when the number of iterations is known or finite.
    - **while Loop**
        - Used when you want to repeat something until a condition is no longer true.
        - Best when the number of iterations is unknown — depends on a condition.

14. Describe a scenario where a while loop is more suitable than a for loop.      
   - A **while** loop is ideal when you do not know in advance how many times you need to repeat something — you're waiting for a condition to become false, rather than iterating over a sequence.  
   - **Example:** Suppose you're writing a program that asks a user to enter a password until they get it right. You don’t know how many attempts the user will need — so a while loop is more appropriate.
   - **Why Not Use a for Loop?**
      - You can’t use a for loop effectively here because:
      - You don’t know how many times the user will get the password wrong.
      - The loop needs to continue until the correct password is entered — that's a condition, not a fixed count.
